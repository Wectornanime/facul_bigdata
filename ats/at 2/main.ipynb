```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('Cancelled_Diverted_2023.csv')
```

# Airline

```python
df_airline = dados.Airline.value_counts()

df_airline
```

```python
plt.barh(df_airline.index.astype(str), df_airline, color='lightblue')
plt.title("Distribuicao por Linha de Vool")
plt.xlabel("Quantidade")
plt.ylabel("Linha de Voo")
plt.show()
```

```python
plt.pie(df_airline)
plt.title("Distribuicao por Linha de Vool")
plt.legend(df_airline.index.astype(str))
plt.show()
```

# Dep Airport (20 maiores)

```python
df_depAirport = dados.Dep_Airport.value_counts() #realiza a contagem dos dados da coluna Dep_Airport
df_depAirport = df_depAirport[:20] #seleciona os 20 primeiros valores
df_depAirport = df_depAirport.sort_values() #ordena os valores de forma crescente
df_depAirport #mostra o resultado
```

```python
plt.barh(df_depAirport.index.astype(str), df_depAirport, color='lightblue')
plt.title("20 maiores aeroportos de destinos")
plt.xlabel("Quantidade")
plt.ylabel("Destinos")
plt.show()
```

# Arr Airport (20 maiores)

```python
df_arrAirport = dados.Arr_Airport.value_counts()

df_arrAirport = df_arrAirport[:20]

df_arrAirport = df_arrAirport.sort_values()

df_arrAirport
```

```python
plt.barh(df_arrAirport.index.astype(str), df_arrAirport, color='lightblue')
plt.title("20 maiores aeroportos de origem")
plt.xlabel("Quantidade")
plt.ylabel("Origem")
plt.show()
```
